```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import itertools

# Load cleaned dataset
df = pd.read_csv("netflix_cleaned.csv") 
```

```python
# Check first few rows
df.head()
```

```python
# Dataset info
df.info()
```

#Data cleaning

```python
# Rename columns
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Drop duplicates
df.drop_duplicates(inplace=True)

# Fill missing values safely (no chained assignment warning)
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('unknown')
df['rating'] = df['rating'].fillna('Not Rated')
df['duration'] = df['duration'].fillna('Unknown')


# Convert 'date_added' to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Standardize text format
df['country'] = df['country'].str.strip().str.lower()
```

#Exploratory Data Analysis (EDA)
1️⃣ Top Countries

```python
top_countries = df['country'].value_counts().head(10)

sns.barplot(x=top_countries.values, y=top_countries.index, palette="Blues_d")
plt.title("Top 10 Countries by Number of Netflix Titles")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

#2️⃣ Top 10 Genres

```python
all_genres = df['listed_in'].dropna().str.split(', ')
flat_genres = list(itertools.chain.from_iterable(all_genres))
top_genres = pd.Series(Counter(flat_genres)).sort_values(ascending=False).head(10)

sns.barplot(x=top_genres.values, y=top_genres.index, palette="Greens_d")
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

#3️⃣ Type Distribution (Movie vs TV Show)

```python
type_counts = df['type'].value_counts()

sns.barplot(x=type_counts.index, y=type_counts.values, palette="Set2")
plt.title("Content Type Distribution")
plt.ylabel("Count")
plt.xlabel("Type")
plt.tight_layout()
plt.show()
```
